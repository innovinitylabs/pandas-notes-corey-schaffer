```python
import pandas as pd
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
dft = pd.DataFrame(people)
```

to get everyone with last name Doe

we get only a `series` with yes or no

```python
dft['last'] == 'Doe'
```

so we are adding it to a variable for clarity and then later we can use it to filter and get the data we need

```python
filt = (dft['last'] == 'Doe') # () added for clarity
```

we can put it directly in the brackets for dataframe  but for clarity we are adding it variable and use

```python
dft[filt]
```

but its better to use `.loc` since we can also filter using column if needed

```python
dft.loc[filt, 'email']
```

#### AND OR Operator in pandas
> AND = `&`  
> OR  = `|`

```python
filt2 = (dft['last'] == 'Doe') & (dft['first'] == 'John')
```

```python
dft.loc[filt2, 'email']
```

to negate the filter we use `~` (kinda like NOT)

```python
dft.loc[~filt2, 'email']
```
